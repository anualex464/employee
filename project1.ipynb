```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv("myexcel.csv")  
df['height'] = np.random.randint(150, 181, size=len(df))


team_count = df['Team'].value_counts()
team_percentage = (team_count / len(df)) * 100

print("1. Team-wise employee count:")
print(team_count)
print("\nPercentage splitting with respect to the total employees:")
print(team_percentage)


position_count = df['Position'].value_counts()
print("\n2. Employee count based on positions:")
print(position_count)


bins = [20, 30, 40, 50, 60, 70]
age_groups = pd.cut(df['Age'], bins=bins)
age_group_counts = age_groups.value_counts()
print("\n3. Employee count in different age groups:")
print(age_group_counts)


highest_salary_team_position = df.groupby(['Team', 'Position'])['Salary'].sum().idxmax()
print("\n4. Team and position with the highest spending in terms of salary:")
print(highest_salary_team_position)


plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Age', y='Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('Correlation between Age and Salary')
plt.show()
```
